```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('/content/Real_Estate_Sales_2001-2022_GL.csv')
```

```python
df.head(2)
```

```python
df.info()
```

```python
df['Date Recorded']=df['Date Recorded'].astype('datetime64[ns]')
```

```python
df['List Year'].agg(['min', 'max','mean','median'])
```

```python
plt.figure(figsize=(10, 6))
sns.set_style("whitegrid")

# Plot the KDE
sns.kdeplot(df['List Year'], fill=True, color="skyblue", linewidth=2)

# Labels and Title
plt.title('Density of Listings by Year', fontsize=16)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Density', fontsize=12)

# Rotate x-axis ticks for clarity
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

```python
df['List Year'].agg(['min', 'max','mean','median'])
df[df['List Year']<2020]
df.drop(df[df['List Year']<2020].index,axis=0,inplace=True)
df['List Year'].value_counts()
```

```python
df["Month"]=df['Date Recorded'].dt.strftime("%B").sort_values(ascending=False)
df["Month"].value_counts()
```

```python
plt.figure(figsize=(10, 6))
sns.set_style("whitegrid")
sns.countplot(data=df, x='Month', palette="viridis")
plt.title('Count of Listings by Month', fontsize=16)
plt.xlabel('Month', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.set_style("whitegrid")
sns.kdeplot(data=df, x='Date Recorded',shade=True,palette="viridis",color="#b78098")
plt.title('Count of Listings by Year', fontsize=16)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
temp1=df["Town"].value_counts().sort_values(ascending=False).head(10).reset_index()
temp1.columns=["Town","Count"]
temp1
```

```python
plt.figure(figsize=(10,6))
sns.set_style("whitegrid")
sns.barplot(data=temp1, y=temp1['Town'],x=temp1['Count'], palette="flare")
plt.title('Count of Listings by Town', fontsize=16)
plt.xlabel('Town', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
df['log_assessed_value'] = np.log1p(df['Assessed Value'])  # log(1 + x)
df['log_assessed_value'].median()

import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
sns.kdeplot(df['log_assessed_value'], fill=True, color="skyblue", linewidth=2)
plt.axvline(df['log_assessed_value'].median(), color='red', linestyle='dashed', linewidth=2)
plt.axvline(df['log_assessed_value'].mean(), color='green', linestyle='dashed', linewidth=1)
plt.title('Density Plot of Log-Transformed Assessed Value', fontsize=16)
plt.xlabel('Log(Assessed Value)', fontsize=12)
plt.ylabel('Density', fontsize=12)
plt.grid(True)
plt.tight_layout()
plt.show()
```
